```python
cleaned_df['Revenue'] = cleaned_df['Quantity'] * cleaned_df['Price']
cleaned_df['Revenue']
```

```python
annulation_filter = cleaned_df[cleaned_df['Invoice'].str.contains('C')]
annulation_filter
```
